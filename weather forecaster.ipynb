```python
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
model=Ridge(alpha=.1)
#model=RandomForestRegressor()
#model=DecisionTreeRegressor()
#model=LinearRegression()
```

```python
features=["PRCP","TMAX","TMIN"]
```

```python
#creating new attributes

core_weather["tempavg_month"]=core_weather["TMAX"].rolling(30).mean()
```

```python
core_weather["Temp_day"]=core_weather["tempavg_month"]/core_weather["TMAX"]
```

```python
core_weather["max_min"]=core_weather["TMAX"]/core_weather["TMIN"]
```

```python
core_weather["monthly_avg"]=core_weather["TMAX"].groupby(core_weather.index.month).apply(lambda x:x.expanding(1).mean())
```

```python
core_weather
```

```python
core_weather["dayof_year_avg"]=core_weather["TMAX"].groupby(core_weather.index.day_of_year).apply(lambda x:x.expanding(1).mean())
```

```python
core_weather
```

```python
features=["PRCP","TMAX","TMIN","tempavg_month","Temp_day","max_min","monthly_avg","dayof_year_avg"]
```

```python
core_weather
```

```python
core_weather=core_weather.iloc[30:,:].copy() #removing null values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
train,test=train_test_split(core_weather,test_size=0.2,random_state=42)
```

```python
model.fit(train[features],train["target"])
```

```python
predictions=model.predict(train[features])
```

```python
#evaluating the model
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(train["target"],predictions)
```

```python
import numpy as np
rmse=np.sqrt(mse)
```

```python
rmse
```

```python
combined=pd.concat([train["target"],pd.Series(predictions,index=train.index)],axis=1)
combined.columns= ["actual","predicted"]
```

```python
# evaluating using cross-validation to avoid overfitting
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,train[features],train["target"],scoring="neg_mean_squared_error")
```

```python
rmse_scores=np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("scores",scores)
    print("mean",scores.mean())
    print('standard deviation',scores.std())
    
```
